```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y,"red")
plt.xlabel("The no. between 5")
plt.ylabel("The square of those no.s")
plt.title("Rajesh ki ghanti")
plt.show()
plt.savefig("rajesh ki ghanti.pdf")
```

## Creating Multiplots on Same Canvas

for this we use **subplot**

```python
# plt.subplot(nrows, ncols, plot_number) 
plt.subplot(1,2,1)
plt.plot(x ,y ,"r--")
plt.title("Ohh yeah")

plt.subplot(1,2,2)
plt.plot(y, x, "blue", linestyle="-.")
plt.title("Kaam 25 hai")
```

# Question practice 

Create a subplot with 1 row and 2 columns (subplot(1,2,1) and subplot(1,2,2)).

In the first subplot:

Plot names vs ages

Use a green dashed line.

Label it "Age Analysis".

Set x-label as "Names" and y-label as "Age".

In the second subplot:

Plot names vs salaries

Use a blue dash-dot line.

Label it "Salary Analysis".

Set x-label as "Names" and y-label as "Salary".

Add a legend for both plots.

Add titles for each plot as "Age Group" and "Salary Group".

Adjust layout using plt.tight_layout().

Save the full figure as "final_practice.pdf".

Finally, display the plot.

```python
names = ['Anya', 'Brian', 'Chris', 'Diana', 'Ethan']
ages = [22, 25, 21, 24, 23]
salaries = [45000, 52000, 43000, 48000, 50000]

```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.subplot(1,2,1)
plt.plot(names, ages, "g--", label = "Age analysis")
plt.xlabel("Names")
plt.ylabel("Ages")
plt.title("Age group")
plt.tight_layout() #to adjust the layout automatically
plt.legend()

plt.subplot(1,2,2)
plt.plot(names, salaries, "b--", label = "Salary analysis")
plt.xlabel("Names")
plt.ylabel("Salaries")
plt.title("Salary Group")
plt.tight_layout() #to adjust the layout automatically
# plt.show()
plt.legend()


plt.savefig("ohhyeah.pdf")
```
